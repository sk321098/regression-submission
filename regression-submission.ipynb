```python
import pandas as pd

literacy_data = pd.read_csv('../Youth Illiteracy Rate (Percentage).csv')
literacy_data.head()
```

```python
literacy_data.describe()
```

```python
year = literacy_data['Unnamed: 0']
literacy_rate = literacy_data['Youth literacy rate, population 15-24 years, both sexes']
import matplotlib.pyplot as plt

plt.figure(figsize = (20,20))
plt.ylabel('Youth literacy rate, population 15-24 years, both sexes')
plt.xlabel('Unnamed: 0')
plt.title('Literacy Rates Over Time')
plt.scatter(year, literacy_rate)
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
x_ary = literacy_data['Unnamed: 0']
x_with_cnst = sm.add_constant(x_ary)
model = sm.OLS(literacy_data['Youth literacy rate, population 15-24 years, both sexes'], x_with_cnst)
results = model.fit()
b,m = results.params
print(m)
print(b)
```

```python
print(results.summary())
```

```python
plt.figure(figsize = (20,20))
plt.scatter(literacy_data['Unnamed: 0'], literacy_data['Youth literacy rate, population 15-24 years, both sexes'])
plt.plot(x_ary, m*x_ary+b, color = 'lawngreen', linewidth = 1)
```
